````python
import streamlit as st
import tensorflow as tf
import numpy as np
import joblib
import pandas as pd

# Load the pre-trained model and scalers
try:
    model = tf.keras.models.load_model('best_lstm_model.keras')
    scaler_X = joblib.load('scaler_X.pkl')
    scaler_y = joblib.load('scaler_y.pkl')
    st.success("Model and scalers loaded successfully!")
except Exception as e:
    st.error(f"Error loading model or scalers: {e}")
    st.stop()

# Streamlit App Title
st.title('Weather Prediction App')
st.write('Predicting \'Next_Day_Rainfall\' based on current weather conditions.')

# Input features from user
st.sidebar.header('Input Weather Parameters')

def user_input_features():
    temperature = st.sidebar.slider('Temperature (Celsius)', -10.0, 50.0, 25.0)
    humidity = st.sidebar.slider('Humidity (%)', 0.0, 100.0, 70.0)
    wind_speed = st.sidebar.slider('Wind Speed (km/h)', 0.0, 100.0, 15.0)
    pressure = st.sidebar.slider('Pressure (hPa)', 900.0, 1100.0, 1012.0)

    data = {
        'Temperature(C)': temperature,
        'Humidity(%)': humidity,
        'WindSpeed(km/h)': wind_speed,
        'Pressure(hPa)': pressure
    }
    features = pd.DataFrame(data, index=[0])
    return features

df_input = user_input_features()

st.subheader('User Input Parameters')
st.write(df_input)

if st.button('Predict Next Day Rainfall'):
    # Preprocess the input data
    # The scaler_X was trained on specific columns, ensure the input matches
    input_array = df_input[['Temperature(C)', 'Humidity(%)', 'WindSpeed(km/h)', 'Pressure(hPa)']].values
    input_scaled = scaler_X.transform(input_array)

    # LSTM expects input in the shape (batch_size, timesteps, features)
    # Here, we assume timesteps=1 for a single prediction
    input_reshaped = input_scaled.reshape(input_scaled.shape[0], 1, input_scaled.shape[1])

    # Make prediction
    prediction_scaled = model.predict(input_reshaped)

    # Inverse transform the prediction to get actual rainfall value
    prediction = scaler_y.inverse_transform(prediction_scaled)

    st.subheader('Prediction: Next Day Rainfall')
    # Assuming the model predicts a single value for rainfall
    predicted_rainfall = prediction[0][0]
    st.metric("Predicted Next Day Rainfall (mm)", f"{predicted_rainfall:.2f}")

    if predicted_rainfall > 0.5:
        st.success("It is likely to rain tomorrow!")
    else:
        st.info("It is unlikely to rain tomorrow.")

# To run this app, save it as `app.py` and run `streamlit run app.py` in your terminal.
# Since we are in Colab, we will write this content to a file and provide instructions.
````

### Step 1: Prepare Data for New Model Training

First, we'll load the original dataset, select the 5 input features you want, and prepare the target variable ('Next_Day_Rainfall'). For this, we'll use `PRECTOTCORR` shifted by one day as the target (predicting tomorrow's rainfall based on today's conditions).

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load the original dataset, skipping the header rows
df_full = pd.read_csv('Kerala_Weather.csv', skiprows=17)

# Rename columns for pd.to_datetime to correctly identify month and day
df_full_renamed = df_full.rename(columns={'MO': 'month', 'DY': 'day'})

# Combine date columns for sorting and time-series operations
df_full['DATE'] = pd.to_datetime(df_full_renamed[['YEAR', 'month', 'day']])
df_full = df_full.sort_values(by='DATE')

# Define the 5 input features you want to use
input_features = ['RH2M', 'T2M', 'WS2M', 'PS', 'PRECTOTCORR']

# Define the target variables:
# 1. Next Day Rainfall (PRECTOTCORR shifted by 1 day)
df_full['Next_Day_Rainfall'] = df_full['PRECTOTCORR'].shift(-1)
# 2. Next Day Temperature at 2 Meters (T2M shifted by 1 day)
df_full['Next_Day_T2M'] = df_full['T2M'].shift(-1)

# Drop rows with NaN values that result from shifting (last row for targets, or any missing values)
df_filtered = df_full[input_features + ['Next_Day_Rainfall', 'Next_Day_T2M']].dropna()

# Separate features (X) and targets (y_rainfall, y_t2m)
X = df_filtered[input_features].values
y_rainfall = df_filtered['Next_Day_Rainfall'].values.reshape(-1, 1) # Reshape for scaler
y_t2m = df_filtered['Next_Day_T2M'].values.reshape(-1, 1) # Reshape for scaler

print(f"Shape of X (features): {X.shape}")
print(f"Shape of y_rainfall (target): {y_rainfall.shape}")
print(f"Shape of y_t2m (target): {y_t2m.shape}")

# Initialize and fit new scaler for X
scaler_X_new = MinMaxScaler()
X_scaled = scaler_X_new.fit_transform(X)

# Initialize and fit new scalers for each target
scaler_y_rainfall_new = MinMaxScaler()
y_rainfall_scaled = scaler_y_rainfall_new.fit_transform(y_rainfall)

scaler_y_t2m_new = MinMaxScaler()
y_t2m_scaled = scaler_y_t2m_new.fit_transform(y_t2m)

print("Data scaled successfully!")

# Save the new scalers
import joblib
joblib.dump(scaler_X_new, 'scaler_X_new.pkl')
joblib.dump(scaler_y_rainfall_new, 'scaler_y_rainfall_new.pkl')
joblib.dump(scaler_y_t2m_new, 'scaler_y_t2m_new.pkl')
print("New scalers (scaler_X_new.pkl, scaler_y_rainfall_new.pkl, scaler_y_t2m_new.pkl) saved successfully!")

# For LSTM, reshape X to (samples, timesteps, features)
# We're using 1 timestep for simplicity as we predict next day from current day
X_reshaped = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

print(f"Shape of X for LSTM: {X_reshaped.shape}")
````

### Step 2: Train a New LSTM Model

Now, let's define and train a new LSTM model using the 5 features. We will then save this new model.

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# --- Model for Rainfall Prediction ---
# Define the LSTM model for rainfall
model_rainfall = Sequential([
    LSTM(50, activation='relu', input_shape=(X_reshaped.shape[1], X_reshaped.shape[2])),
    Dense(1) # Output a single value for rainfall
])

model_rainfall.compile(optimizer='adam', loss='mse')

print("LSTM model for Rainfall created and compiled.")

# Train the rainfall model
history_rainfall = model_rainfall.fit(X_reshaped, y_rainfall_scaled, epochs=50, batch_size=32, verbose=0)

print("LSTM model for Rainfall trained successfully.")

# Save the rainfall model
model_rainfall.save('best_lstm_model_rainfall.keras')
print("Rainfall model (best_lstm_model_rainfall.keras) saved successfully!")

# --- Model for Temperature Prediction ---
# Define the LSTM model for temperature
model_temperature = Sequential([
    LSTM(50, activation='relu', input_shape=(X_reshaped.shape[1], X_reshaped.shape[2])),
    Dense(1) # Output a single value for temperature
])

model_temperature.compile(optimizer='adam', loss='mse')

print("LSTM model for Temperature created and compiled.")

# Train the temperature model
history_temperature = model_temperature.fit(X_reshaped, y_t2m_scaled, epochs=50, batch_size=32, verbose=0)

print("LSTM model for Temperature trained successfully.")

# Save the temperature model
model_temperature.save('best_lstm_model_temperature.keras')
print("Temperature model (best_lstm_model_temperature.keras) saved successfully!")
````

### Step 3: Update `app.py` for New Model and Inputs

Finally, we need to update your `app.py` file to load the newly trained model and scalers, and to accept only the 5 specific input features from the user.

````python
with open('app.py', 'w') as f:
    f.write("""import streamlit as st
import tensorflow as tf
import numpy as np
import joblib
import pandas as pd

# Load the new pre-trained model and scalers
try:
    model_rainfall = tf.keras.models.load_model('best_lstm_model_rainfall.keras')
    model_temperature = tf.keras.models.load_model('best_lstm_model_temperature.keras')
    scaler_X = joblib.load('scaler_X_new.pkl')
    scaler_y_rainfall = joblib.load('scaler_y_rainfall_new.pkl')
    scaler_y_t2m = joblib.load('scaler_y_t2m_new.pkl')
    st.success("New models and scalers loaded successfully!")
except Exception as e:
    st.error(f"Error loading new models or scalers: {e}")
    st.stop()

# Streamlit App Title
st.title('Weather Prediction App')
st.write("Predicting 'Next Day Rainfall' and 'Next Day Temperature' based on current weather conditions (5 features).")

# Input features from user
st.sidebar.header('Input Weather Parameters')

def user_input_features():
    # Note: Units are based on the original dataset description
    rh2m = st.sidebar.slider('Relative Humidity at 2 Meters (%) (RH2M)', 0.0, 100.0, 70.0)
    t2m = st.sidebar.slider('Temperature at 2 Meters (C) (T2M)', -10.0, 50.0, 25.0)
    ws2m = st.sidebar.slider('Wind Speed at 2 Meters (m/s) (WS2M)', 0.0, 20.0, 5.0) # Original data is m/s
    ps = st.sidebar.slider('Surface Pressure (kPa) (PS)', 90.0, 110.0, 100.0) # Original data is kPa
    prectotcorr = st.sidebar.slider('Current Day Precipitation (mm/day) (PRECTOTCORR)', 0.0, 50.0, 5.0)

    data = {
        'RH2M': rh2m,
        'T2M': t2m,
        'WS2M': ws2m,
        'PS': ps,
        'PRECTOTCORR': prectotcorr
    }
    features = pd.DataFrame(data, index=[0])
    return features

df_input = user_input_features()

st.subheader('User Input Parameters')
st.write(df_input)

if st.button('Predict Next Day Weather'):
    # Preprocess the input data using the new scaler
    input_array = df_input[['RH2M', 'T2M', 'WS2M', 'PS', 'PRECTOTCORR']].values
    input_scaled = scaler_X.transform(input_array)

    # LSTM expects input in the shape (batch_size, timesteps, features)
    input_reshaped = input_scaled.reshape(input_scaled.shape[0], 1, input_scaled.shape[1])

    # Make predictions for Rainfall
    prediction_rainfall_scaled = model_rainfall.predict(input_reshaped)
    predicted_rainfall = scaler_y_rainfall.inverse_transform(prediction_rainfall_scaled)[0][0]

    # Make predictions for Temperature
    prediction_t2m_scaled = model_temperature.predict(input_reshaped)
    predicted_t2m = scaler_y_t2m.inverse_transform(prediction_t2m_scaled)[0][0]

    st.subheader('Prediction: Next Day Rainfall & Temperature')
    st.metric("Predicted Next Day Rainfall (mm)", f"{predicted_rainfall:.2f}")
    st.metric("Predicted Next Day Temperature (C)", f"{predicted_t2m:.2f}")

    if predicted_rainfall > 0.5:
        st.success("It is likely to rain tomorrow!")
    else:
        st.info("It is unlikely to rain tomorrow.")
""")

print("Streamlit app.py updated for multiple predictions successfully!")
````
