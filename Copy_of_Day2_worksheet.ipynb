```python
df=df.drop(columns=['deck'])
df.shape
```

```python
df.dropna(subset=['embarked','embark_town'],inplace=True)
print(df.isnull().sum())
```

```python
age_mean=df['age'].mean()
df['age']=df['age'].fillna(age_mean)
```

```python
#creating new column titled as age_group
df['age_group']=np.where(df['age']<18,"Child","Adult")
#creating new column titled as fare_per_person
df['family_size']=df['sibsp']+df['parch']
df['fare_per_person']=df['fare']/(df['family_size']+1)
```

```python
df.info()
```

```python
#renaming a column
df.rename(columns={'pclass':'Passenger_Class'},inplace=True)
df
```

```python
#filtering and conditioning
df[df['age']>df['age'].mean()]
```

```python
df[(df['sex']=='female')&(df['Passenger_Class']==1)]
```

```python
#survival rate by gendder
df.groupby('sex')['survived'].mean()
```

```python
#avg fare by class
df.groupby('class')['fare'].mean()
```

```python
#correlation
df.corr(numeric_only=True)
```
