```python
from sklearn.preprocessing import OneHotEncoder
OneHotEncoder()
df=pd.get_dummies(df,columns=['cut','color','clarity'],prefix=['CUT','COLOR','CLARITY'],drop_first=True)           #drop_first is used to drop the column titled embark
df.head()
```

```python
bool_col=df.select_dtypes(include='bool').columns
df[bool_col]=df[bool_col].astype(int)
df
```

```python
#scaling using winsorize
from scipy.stats.mstats import winsorize

#applying winsorize method to carat columns

for col in ['carat','depth','table']:
  df[col]=winsorize(df[col],limits=(0.05,0.05))

df.head()

```

```python
y=df['price']
```

```python
x=df.drop(['price'],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print(f'mae:{mean_absolute_error(y_test,y_pred)}')
print(f'mse:{mean_squared_error(y_test,y_pred)}')
print(f'2:{r2_score(y_test,y_pred)}')
```
